```python
import polars as pl
```

```python
pl.scan_ndjson('data/*.jsonl', include_file_paths='date').head().collect()
```

```python
df = (
    pl.scan_ndjson('data/*.jsonl', include_file_paths='date')
     .with_columns(
         date=pl.col("date").str.replace("data/", "").str.replace(".jsonl", "").str.to_datetime("%Y-%m-%d"),
         growth=pl.col("total_downloads") / pl.col("total_downloads").min().over("full_name")
     )
     .filter(pl.col('total_downloads') > 20000)
     .collect()
)
```

```python
import altair as alt
alt.data_transformers.enable("vegafusion")

alt.Chart(df).mark_line().encode(x='date', y='open_issues:Q', color='full_name', tooltip=['full_name']).interactive().properties(width=700)
```
